```python
# убираем эффект довнесения средств
daily_change_fixed = daily_change

for i in range(len(daily_change_fixed) - 1):
    if daily_change_fixed[i] > 8:
        daily_change_fixed[i] = np.mean(daily_change_fixed)
        
for i in range(len(daily_change_fixed) - 1):
    if i<100 and daily_change_fixed[i]>7:
        daily_change_fixed[i] = np.mean(daily_change_fixed)
```
